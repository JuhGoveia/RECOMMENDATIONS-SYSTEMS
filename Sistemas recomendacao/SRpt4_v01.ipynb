## Criamos um Sistema de Recomendação baseado em filtragem colaborativa baseada nas avaliações dos clientes nos seus produtos adquiridos.

```python
from sklearn.metrics.pairwise import cosine_similarity, pairwise_distances
import pandas as pd
```

```python
# Criar um DataFrame de exemplo com avaliações de usuários para diferentes itens
data = {
    'Usuario': ['Kaio', 'Evandro', 'Julia', 'Pedro'],
    'Poupanca': [5, 4, 0, 2],
    'Investimentos': [2, 5, 4, 0],
    'Emprestimo': [0, 3, 5, 4],
    'Cheque_especial': [4, 0, 2, 5]
}

df = pd.DataFrame(data)
df
```

```python
# Definir o usuário-alvo e os itens avaliados por esse usuário para receber uma recomendação de produto

usuario_alvo = 'Kaio'
itens_avaliados = {'Poupanca': 2, 'Investimentos': 1, 'Emprestimo': 4, 'Cheque_especial': 5}
```

```python
# Calcular a similaridade de coseno entre o usuário-alvo e os outros usuários
similaridade = 1 - pairwise_distances(df.set_index('Usuario').loc[usuario_alvo].values.reshape(1, -1), df.set_index('Usuario').values, metric='cosine')
```

```python
# Criar um DataFrame com as similaridades calculadas
similaridades_df = pd.DataFrame(similaridade, columns=df['Usuario'])
similaridades_df
```

```python
# Identifica o cliente mais semelhante diferente dele proprio
cliente_mais_semelhante = similaridades_df.loc[0][similaridades_df.loc[0] < 1].idxmax()
cliente_mais_semelhante
```

```python
#identificando os produtos do cliente mais semelhante e recomendando o de maior valor avaliado

produtos_cliente_semelhanate_df = df[df['Usuario']==cliente_mais_semelhante]

produto_recomendado = produtos_cliente_semelhanate_df.iloc[:, 1:].idxmax(axis=1).values[0]
produto_recomendado
```

```python
print(f"O Produto recomendado para {usuario_alvo} é o {produto_recomendado}, devido a sua grande semelhança com {cliente_mais_semelhante}.")
```
